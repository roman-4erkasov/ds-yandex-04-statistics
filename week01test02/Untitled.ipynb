```python
df = pd.read_csv("./pines.txt", sep="\t")
```

```python
df.shape
```

```python
boundaries = [(0,40),(40,80),(80,120),(120,160),(160,201)]
groups = []
```

```python
# list(itertools.product(boundaries,boundaries))
```

```python
for sn, we in list(itertools.product(boundaries,boundaries)):
#     print(sn,we)
    groups.append(
        df[
            df.sn.apply(lambda x:sn[0]<=x<sn[1]) 
            & 
            df.we.apply(lambda x:we[0]<=x<we[1])
        ].shape
    )
np.mean(groups)
```
