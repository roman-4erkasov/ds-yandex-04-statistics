```python
import pandas as pd
from scipy import stats
import numpy as np
```

```python
df = pd.read_csv("./water.txt", sep='\t')

df
```

```python
from statsmodels.stats.weightstats import _zconfint_generic, _tconfint_generic

df1 = df.loc[df.location=='North', "mortality"]

# print _zconfint_generic(
#     df1.mean(), 
#     df1.std(ddof=1)/df1.shape[0]**0.5,
#     0.05,
#     'two-sided'
# )

# int_len = stats.norm.ppf(1-0.025)*df1.std(ddof=1)/df1.shape[0]**0.5
# print df1.mean() - int_len, df1.mean() + int_len

print _tconfint_generic(
    df1.mean(),
    df1.std(ddof=1)/df1.shape[0]**0.5,
    df1.shape[0] - 1,
    0.05,
    'two-sided'
)

int_len = stats.t.ppf(1-.05/2, df1.shape[0] - 1)*df1.std(ddof=1)/df1.shape[0]**0.5
print df1.mean() - int_len, df1.mean() + int_len
```

```python
import math
df2 = df.loc[df.location=='South', "mortality"]

# int_len = stats.norm.ppf(1-0.025)*df2.std(ddof=1)/df2.shape[0]**0.5
# print df2.mean() - int_len, df2.mean() + int_len

# print _zconfint_generic(
#     df2.mean(), 
#     df2.std(ddof=1)/df2.shape[0]**0.5,
#     0.05,
#     'two-sided'
# )

print _tconfint_generic(
    df2.mean(), 
    df2.std(ddof=1)/math.sqrt(df2.shape[0]),#**0.5,
    df2.shape[0] - 1,
    0.05,
    'two-sided'
)

int_len = stats.t.ppf(1-.05/2, df2.shape[0] - 1)*df2.std(ddof=1)/df2.shape[0]**0.5
print "%.4f"%(df2.mean() - int_len), df2.mean() + int_len
```
